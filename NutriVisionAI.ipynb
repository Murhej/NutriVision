Get the Data


```python
import kagglehub
import os

# Download dataset
path = kagglehub.dataset_download("dansbecker/food-101")

print("Path to dataset files:", path)
print(os.listdir(path))
print("----------------------------------")

# Define dataset paths
food_image_path = os.path.join(path, "food-101", "food-101", "images")
food_meta_path = os.path.join(path, "food-101", "food-101", "meta")

print("Images Path:", food_image_path)
print("Meta Path:", food_meta_path)
print("----------------------------------")

# List folders
folder_image = os.listdir(food_image_path)
folder_meta = os.listdir(food_meta_path)

# Show food classes
print("Food classes:")
for f in folder_image:
    if f != "__MACOSX":
        print(f)

print("----------------------------------")

# Show meta files
print("Files inside meta:")
for f in folder_meta:
    if f != "__MACOSX":
        print(f)
```

### **COMMENT ON THE CODE TOP**
In the Get the Data step, we downloaded a large dataset that contains many images and metadata files used for training the model.

We ignored the __MACOSX folder because it is created for Mac operating systems and is not needed when working on Windows. It only contains system metadata. We filtered it out so the program only shows useful dataset files.

After that, we created loops to display all folders inside the images directory and list the files inside the meta directory.
